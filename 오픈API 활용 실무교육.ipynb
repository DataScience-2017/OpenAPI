```python
# html에서 데이터 추출하기 : BeautifulSoup

from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(data2, "html.parser")
```

```python
soup
```

```python
docs = soup.find_all('doc')
print(len(docs))
```

```python
docs[0]
```

```python
# find = 하나하나에서 태그의 위치를 찾음
# find_all = 전체어서 찾음
# get_text = 하나의 태그 안의 값을 반환해줘
docs[1].find('bookname').get_text()
```

```python
docstxt = []

for i in docs:
    ranking = i.find("ranking").get_text()
    bookname = i.find("bookname").get_text()
    authors = i.find("authors").get_text()
    publisher = i.find("publisher").get_text()
    publication_year = i.find("publication_year").get_text()
    isbn13 = i.find("isbn13").get_text()
    
    docstxt.append({"순위" : ranking, "책이름": bookname,"작가명" :authors,
                    "출판사명":publisher, "발행년도" : publication_year,
                    "ISBN" :isbn13})
    
import pandas as pd

df1 = pd.DataFrame(docstxt)
df1
```
